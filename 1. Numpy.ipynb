```
# 행렬 연산 함수
a = np.array([[1,2],[3,4]])
s = np.sum(a)
s = np.sum(a, axis = 0) # axis 0 = 열. 열끼리만 더함
print(s)
s = np.sum(a, axis = 1) # axis 1 = 행. 행끼리만 더함
print(s)


# 차원의 변환
Y1 = np.arange(12)
print(Y1)
Y2=Y1.reshape(2, 3, 2, order= 'C') # np.reshape(matrix수, row, column, order=‘C’ or ‘F’)
print(Y2)
Y2=Y1.reshape(3, 2, 2, order= 'C')
print(Y2)
# order=‘C’: 값을 row부터 채움(기본값, 생략가능)
# order=‘F’: 값을 column부터 채움

Y3=np.arange(3.5, 10.5, 2, dtype=float)   
print(Y3)
print(np.amax(Y3)) # 최대값 반환
print(Y3.dtype) # Y3의 데이터 타입

Y6 = np.arange(1,5).reshape((2,2))
print(Y6)
print(np.amin(Y6))  # 최소값 1을 반환. amin or amax
print(Y6.shape)     # Y6의 행렬 구조 반환
```

```
# boolean indexing

# names의 각 요소가 data의 각 column에 마스크(적용) 를 하고, names이 BTS인 행의 모든 data만을 보고자 할때
names = np.array(['BTS', 'SongGaIn','HongJo','ChoYongPil','GirlGroup'])
number = np.array(['1', '2', '3', '4'])

# 아래에서 사용되는 np.random.randn() 함수는 기대값이 0이고, 표준편차가 1인 가우시안 정규 분포를 따르는 난수를 발생시키는 함수이다.
# 이 외에도 0~1의 난수를 발생시키는 np.random.rand() 함수도 존재한다.
data = np.random.randn(4,5)
print(data)

# 요소가 BTS인 항목에 대한 mask 생성
names_BTS_mask = (names == 'BTS')
names_SongGaIn_mask = (names == 'SongGaIn')
number_4_mask = (number == '4')

# column요소가 4인 항목에 대한 mask 생성
data[number_4_mask,:]

# row요소가 BTS인 항목에 대한 mask 생성
data[:, names_BTS_mask]

# 논리 연산을 응용하여, 요소가 BTS 또는 GirlGroup인 column의 데이터만 출력하기
data[:,(names == 'BTS') | (names == 'GirlGroup')]

# number == '4＇이고 names == 'BTS’에 해당하는 요소들을 찾아낸다. 
data[(number == '4'), (names == 'BTS')]  

# number == ‘3’ 에 해당하는 요소들을 찾아낸다. 
data[(number=='3'),:]

#위에서 생성한 data에서 모든  row 중에서 0번 column(열)의 값이 0보다 작은 row(행)을 구한다.
data[data[:, 0]<0, : ] 

# 0번째 열(column)의 값이 0보다 작은 행(row)의 1,2번 열 값
data[data[:,0]<0, 1:3]

# 0번 column의 값이 0보다 작은 row의 1,2번에 1.0을 대입한다.
data[data[:,0]<0,  1:3] = 1.0
data[data[:,0]<0, :]

# data array에서 1번 column이 1보다 작은 요소의 boolean 값은 다음과 같다.
print(data[:,1])
data[:,1] < 1
```

```
import numpy as np
print(np.random.randint(16)) # 0부터 15까지 중 랜덤 숫자 1개 발생  
print(np.random.rand(5, 4))  # 0~1 사이숫자 [5,4]배열로 발생
print(np.random.randn(5, 4)) # [5,4]배열에 난수 발생, 평균 0, 표준편차 1

a=np.random.randn(5, 4)      # [5,4]배열에 난수 발생, 평균 0, 표준편차 1
b=np.abs(a)                  # 각 요소의 절대값을 계산
c=np.sqrt(a)                 # = a**0.5, 각요소의 제곱근 계산
d=np.exp(a)                  # 각 요소의 무리수 e의 지수 값을 구하는 함수

f=np.log(a)                  # 각 성분을 자연로그 값을 계산하기
g=np.log10(a)                # 상용로그 값 계산, import math 사용시 math.log10(i)형태
h=np.log2(a)                 # 밑이 2인 로그값 계산

print("a")
print(a)
print("b")

print(b)
print("c")

print(c)
print("c")

print(d)
print("d")

print(f)
print("f")

print(g)
print("G")

print(h)
```

```
# argmin, argmax, cumsum, cumprod함수

a= [[-2.0966307,1.27503654,   0.11059781],
      [-0.5951133, 1.23189083,  -0.2504 ]] 

# 전체 성분의 합을 계산 
np.sum(a)

# axis=0, column의 합을 계산, axis=1, row의 합을 계산, 
np.sum(a, axis=0)

# axis=0, column의 평균을 계산
np.mean(a, axis=0)

# 전체 요소의 평균을 계산
np.mean(a)
# 전체 성분의 표준편차, 분산, 최소값, 최대값 계산(std, var, min, max)
np.std(a)
np.min(a, axis=0) # 각 column의 값 중 최소값을 구한다

# axis=0, column 중에서 계산, axis=1, row 중에서 계산, 
np.argmin(a, axis=0)

# axis=0, column 중에서 계산, axis=1, row 중에서 계산 
np.argmax(a, axis=0)

# 맨 처음 성분부터 각 성분까지의 누적합 또는 누적곱을 계산 (cumsum, cumprod)
np.cumsum(a)    # 누적합
np.cumprod(a)   # 누적곱
```

```
# sort 
a = [[3,5,1,2],
     [4,6,2,5],
     [9,0,1,4]]

# axis=0이면 각 column 대해서 오름차순으로 정렬 
print(np.sort(a, axis=0))

# axis=1이면 각 row 대해서 오름차순으로 정렬 
print(np.sort(a, axis=1))

print("a=\n", np.sort(a)[:: 1] ) #  각 row에 대해서 오름차순으로 정리 
```

```
# 행렬 iterate 
#     행렬의 원소를 처음부터 끝까지 순차적으로 접근 혹은 읽어 드림
#     C++, Java iterate처럼 next()메소드를 통해 행렬의 요소에 접근가능 

import numpy as np

a = np.array([[10, 20, 30, 40], 
              [50, 60, 70, 80]])

for x in np.nditer(a, order='F'):
       print(x, end=' ')

print()
for x in np.nditer(a, order='C'):
       print(x, end=' ')

# order=‘F’: 값을 column(열)부터 채움
# order=‘C’: 값을 row(행)부터 채움

print()
it = np.nditer(a, flags=['f_index']) # 인덱스 순서가 열 기준
while not it.finished:
     print("%d(%d)" % (it[0], it.index), end=' ')
     it.iternext()

print()
it = np.nditer(a, flags=['c_index']) # 인덱스 순서가 행 기준
while not it.finished:
     print("%d(%d)" % (it[0], it.index), end=' ')
     it.iternext()

print()
it = np.nditer(a, flags=['multi_index']) # 인덱스 순서가 행,열
while not it.finished:
     print("%d%s" % (it[0], it.multi_index), end=' ')
     it.iternext()
```
